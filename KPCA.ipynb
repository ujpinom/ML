```python
from sklearn.datasets import make_moons;import matplotlib.pyplot as plt;import seaborn as sns;sns.set()


X,y=make_moons(random_state=123)
plt.scatter(X[y==0,0],X[y==0,1])
plt.scatter(X[y==1,0],X[y==1,1], color='red')
```

### Linealizar mediante KPCA un conjunto de datos no separables linealmente

```python
from KPCA import rbf_kernel
X_kpca=rbf_kernel(X,gamma=15,n_components=2)
plt.scatter(X_kpca[y==0,0],X_kpca[y==0,1],color='red',marker='x')
plt.scatter(X_kpca[y==1,0],X_kpca[y==1,1],color='blue',marker='^')
plt.xlabel('KPCA 1');plt.ylabel('KPCA 2')
```

```python
from sklearn.datasets import make_circles

X_cle,y_cle=make_circles(n_samples=1000,random_state=123,factor=0.2,noise=0.1)
plt.scatter(X_cle[y_cle==0,0],X_cle[y_cle==0,1],color='orange',marker='x',alpha=0.6)
plt.scatter(X_cle[y_cle==1,0],X_cle[y_cle==1,1],color='green',marker='o',alpha=0.6)
```

```python
X_cle_kpca=X_kpca=rbf_kernel(X_cle,gamma=15,n_components=2)
plt.scatter(X_cle_kpca[y_cle==0,0],X_cle_kpca[y_cle==0,1],color='orange',marker='x',alpha=0.6)
plt.scatter(X_cle_kpca[y_cle==1,0],X_cle_kpca[y_cle==1,1],color='green',marker='o',alpha=0.6)
plt.xlabel('KPCA 1');plt.ylabel('KPCA 2')
```
