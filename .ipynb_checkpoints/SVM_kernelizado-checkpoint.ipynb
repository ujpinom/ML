#                                       NON-LINEAR KERNEL SVM

```python
import numpy as np;import matplotlib.pyplot as plt;import seaborn as sns;sns.set()
generador=np.random.RandomState(1)
x=generador.randn(200,2)
y=np.logical_xor(x[:,0]>0,x[:,1]>0)
y=np.where(y,1,-1)
plt.scatter(x[y==1,0],x[y==1,1],color='red',marker='x')
plt.scatter(x[y==-1,0],x[y==-1,1],color='blue',marker='o')
```

```python
from sklearn.svm import SVC
modelo=SVC(random_state=1,gamma=0.5,C=100)
modelo.fit(x,y)

```

Altos valores de C hacen que los limites de decisión sean más complejos, por lo que el modelo se adapta bien al conjunto de 
datos de entrenamiento, sin embargo, lo más probable es que dicho algoritmo sufra de overfitting y no clasifique de forma 
correcta conjuntos de datos desconocidos.

```python
from fronteras import visualizador
visualizador.visualizar_fronteras(x,y,modelo)
plt.title('SVM_Kernel=RBF')
```
